```python
   
def batch(k1):    
    
    t = 25
    
    t_span = np.linspace(0,t,100)
    
    N0 = 1
    P = 300
    R = 8.314
    T = 150 + 273.15
    
    V0 = R*T*N0/P
    
    initial_conditions = [0.75,0,0.25,V0]
    
   
    m = 1
    
    
    def rates(t,y):
        
        NA,NB,NC,V = y[0],y[1],y[2],y[3]
        CA,CB,CC = NA/V , NB/V , NC/V
        
        rA = -k1*(CA**m) + (k1/0.061)*(CB**3)
        rB = -3*rA
        rC = 0
        
        dNA = rA*V
        dNB = rB*V
        dNC = 0
        dV = R*T/P*(dNA+dNB+dNC)
    
        return dNA,dNB,dNC,dV
    

    
    solution = solve_ivp(rates,(0,t), initial_conditions,dense_output = True)
    
    sol = solution.sol(t_span)
    
    N = sol[0] + sol[1] + sol[2]
    
    V_V0 = N/N0
    
    y,x = N/N0, t_span
    
    plt.figure(1)
    plt.plot(x,y)
    plt.ylabel('V') 
    plt.xlabel('t') 
    
    plt.figure(1)
    plt.plot(A[:,0],A[:,1])
    plt.ylabel('V') 
    plt.xlabel('t') 

    
    return 1

print(batch(0.13))
```
