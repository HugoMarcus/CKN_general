```python
import numpy as np
from scipy.integrate import solve_ivp
from scipy.optimize import minimize

# Define your experimental data (t and y values)
experimental_t = x  # Fill in your experimental time data
experimental_y = y  # Fill in your experimental y data

beg = x[0]
end = x[-1]
no = len(x)
t_span_c = np.linspace(beg,end,no)

# Define a function to compute the objective (error) between experimental and model data
def objective(params):
    k, n, m = params
    model_y = batch(k, n, m)  # Use your existing batch function to generate model predictions
    error = np.sum((model_y - experimental_y) ** 2)  # Use sum of squared differences as the error
    return error


initial_guess = [20, 1, 1] 

bounds = [(-0.0001,1000), (1,3),(1,3)]

result = minimize(objective, initial_guess, method='Nelder-Mead',bounds = bounds)  

# Get the optimized parameters
optimized_k, optimized_n, optimized_m = result.x

# Print the optimized parameters
print(f'Optimized k: {optimized_k}')
print(f'Optimized n: {optimized_n}')
print(f'Optimized m: {optimized_m}')

print(objective([optimized_k[0], 2, 1 ]))
```
