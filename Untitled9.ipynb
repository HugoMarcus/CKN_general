```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
def CSTR(X_max):
    
    k = 0.5/60
    Q = 4

    CA0 = 3
    CB0 = 0.2
    
    FA0 = CA0*Q
    FB0 = CB0*Q
    
    
    w0 = 0.01
    
    num_i = 100001
    X = np.linspace(0.656707,X_max,num_i)
    
    def rates(FA,FB):
        
        
        CA,CB = FA/Q , FB/Q
        
        rA = -k*CA**1*CB**2
        rB1 = 2*rA
        rB2 = -4*rA
        rB = rB1 + rB2
        rC = 0
        
       
        return rA,rB,rC
    
    
    def solve(var):
        
        FB,V_i = var
        
        rA, rB = rates(FA,FB)[0],rates(FA,FB)[1]
        
        eq1 = FA0 - FA + rA*V_i
        eq2 = FB0 - FB + rB*V_i
        
        return eq1,eq2
    
    X_l = []
    y = []
    V_PFR = []
    V_PFR_i = 0
    W = []
    
    
    for i in range (0,num_i):
        
        if i == 0:
                 
            X_i = X[i]
        
            FA = (1-X_i)*FA0
        
            FB,w_i_n = fsolve(solve,[FB0,w0])[0],fsolve(solve,[FB0,w0])[1]
            
            W.append(w_i_n)
            
            rA = rates(FA,FB)[0]
        
            X_n = X_i
            y_n = FA0/(-rA)
        
            X_l.append(X_n)
            y.append(y_n)
            
            if X_n == 0:
                V_PFR.append(0)
            
            if X_n > 0:
            
                V_n = (X[-1]-X[-2])*y[-1]
            
                V_PFR_i = V_n + V_PFR_i
        
                V_PFR.append(V_PFR_i)
            
            
            
        
        
        if i > 0:
            
            
            X_i = X[i]


            FA = (1-X_i)*FA0


            FB,w_i_n = (fsolve(solve,[10,W[-1]])[0],fsolve(solve,[10,W[-1]])[1])

            W.append(w_i_n)

            rA = rates(FA,FB)[0]

            X_n = X_i
            y_n = FA0/(-rA)

            X_l.append(X_n)
            y.append(y_n)
            
           # if y[-1]-y[-2] >0 :
                
               # print(X_l[-1])
                #print(i)
              
            if X_n == 0:
                V_PFR.append(0)
            
            if X_n > 0:
            
                V_n = (X[-1]-X[-2])*y[-1]
            
                V_PFR_i = V_n + V_PFR_i
            
                V_PFR.append(V_PFR_i)
            

        
        
    plt.figure(1)
    plt.plot(X,y)
    plt.ylabel('FA0/-rA') 
    plt.xlabel('X') 
     
      
    plt.figure(2)
    plt.plot(X,V_PFR,label = 'V_PFR')
    plt.plot(X,W,label = 'V_CSTR') 
    plt.xlabel('X') 
    plt.ylabel('V (l)')
    plt.legend()
    plt.show()
    
    print(V_PFR[100000])
 


print(CSTR(0.99))
```
